```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
df = pandas.read_csv('genshin_weapons_v1.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.columns = ['df.columns','type','rarity', 'base_atk','max_atk','substat_type','base_substat', 'max_substat', 'passive_name','passive_ability','ascension_material_1', 'ascension_material_2','ascension_material_3', 'ascension_material_4','elite_enemy_drop_1', 'elite_enemy_drop_2', 'elite_enemy_drop_3','general_enemy_drop_1', 'general_enemy_drop_2', 'general_enemy_drop_3']
```

```python
len(df)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
plt.style.use('seaborn-whitegrid')

plt.scatter(df['base_atk'], df['max_atk'])

plt.title('Scatter plot of base attack of weapons against their max attack')
plt.xlabel('Base Attack')
plt.ylabel('Max Attack')

plt.show()
```

```python
df.groupby(['type','rarity','elite_enemy_drop_1']).mean()
```

```python
rarity_counts = df['rarity'].value_counts()

plt.pie(rarity_counts.values, labels = rarity_counts.index, autopct='%1.1f%%')

plt.title('Weapons grouped by rarity')

plt.show()
```

```python
mean_maxatk_by_type = df.groupby('type')['max_atk'].mean()

plt.barh(mean_maxatk_by_type.index, mean_maxatk_by_type.values)

for index, value in enumerate(mean_maxatk_by_type.values):
    plt.text(value + 1, index, f"{value:.2f}", va = 'center')

plt.title('Average maximum attack of weapons grouped by weapon type')
plt.xlabel('Average max attack')
plt.ylabel('Type of weapon', labelpad = 10)

plt.show()
```

```python
df.groupby(['type','base_atk','passive_ability']).mean()
```

```python
mean_baseatk_by_type = df.groupby('type')['base_atk'].mean()
plt.barh(mean_baseatk_by_type.index, mean_baseatk_by_type.values)
for index, value in enumerate(mean_baseatk_by_type.values):
    plt.text(value + 1, index, f"{value:.2f}", va = 'center')
plt.title('Average base attack of weapons grouped by weapon type')
plt.xlabel('Average base attack')
plt.ylabel('Type of weapon', labelpad = 10)
plt.show()
```
